```python
df_titleAkas_FR = df_titleAkas[df_titleAkas['region'] == 'FR']   
```

```python
df_titleAkas_langfr = df_titleAkas[df_titleAkas['language'] == 'fr']   
```

```python
df_titleAkas_Original = df_titleAkas[df_titleAkas['isOriginalTitle'] == 1]
```

```python
df_titleAkas_New = pd.concat([df_titleAkas_FR, df_titleAkas_langfr, df_titleAkas_Original], axis = 0)
```

```python
df_titleAkas_New = df_titleAkas_New.drop_duplicates(subset=['titleId', 'ordering', 'title'], keep='first')
```

```python
df_titleAkas_New = df_titleAkas_New.drop(['ordering','types','attributes','isOriginalTitle'],axis = 1)
```
